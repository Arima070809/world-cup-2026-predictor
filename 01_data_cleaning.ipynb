```python
df = df.dropna(subset=['home_score', 'away_score'])
print(df.shape)
```

```python
df = df[df['date'] >= '1990-01-01'].copy()
print(df.shape)
```

```python
df['home_score'] = df['home_score'].astype(int)
df['away_score'] = df['away_score'].astype(int)
df[['home_score', 'away_score']].head()
```

```python
def get_weight(tournament):
    weights = [
        ('FIFA World Cup qualification', 3.0),
        ('FIFA World Cup', 5.0),
        ('UEFA Euro qualification', 2.5),
        ('UEFA Euro', 4.0),
        ('Copa América qualification', 2.0),
        ('Copa América', 4.0),
        ('AFC Asian Cup qualification', 1.5),
        ('AFC Asian Cup', 3.0),
        ('African Cup of Nations qualification', 1.5),
        ('African Cup of Nations', 3.0),
        ('CONCACAF Gold Cup', 3.0),
        ('UEFA Nations League', 2.5),
        ('CONCACAF Nations League', 2.0),
        ('Friendly', 0.5),
    ]
    for key, weight in weights:
        if key.lower() in tournament.lower():
            return weight
    return 1.0

df['tournament_weight'] = df['tournament'].apply(get_weight)
df.to_csv(r"C:\Users\aryan\OneDrive\Desktop\World Cup Predictor\results_clean.csv", index=False)
print("Saved with updated weights!")
```

```python
def get_result(row):
    if row['home_score'] > row['away_score']:
        return 'H'
    elif row['home_score'] < row['away_score']:
        return 'A'
    else:
        return 'D'

df['result'] = df.apply(get_result, axis=1)
df['result'].value_counts()
```
